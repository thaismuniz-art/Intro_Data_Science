## Variáveis

* O nome da variável deve começar com uma letra ou underscore (_), nunca um número
* O nome da variável só pode conter caracteres alfanuméricos e underscore (A-z, 0-9, e _ )
* Nomes de variáveis são sensíveis a minúsculas e maiúsculas (carro, Carro e CARRO são três variáveis diferentes)
* Variáveis não podem ser palavras-chave do Python (https://docs.python.org/3/reference/lexical_analysis.html#keywords)

```python
idade=22
idade_meses=idade*12
```

```python
print("Minha idade é", idade,"anos. Ou seja,",idade_meses,"meses")
```

```python
print(f"Minha idade é {idade} anos. Ou seja, {idade_meses} meses")
```

## Listas
Criando listas

```python
pizza=["calabresa","muçarela","atum",
       "portuguesa","vegana","marguerita"]
```

```python
print(pizza)
print(pizza[-3])
```

Alguns métodos de listas (append(),clear(),copy(), count(), extend(), index(), insert(),pop(), remove(), reverse(), sort())

```python
pizza.append("bode")
print(pizza)
pizza.append("strogonoff")
print(pizza)
```

```python
pizza.remove("bode")
print(pizza)
pizza.reverse()
print(pizza)
```

```python
pizza.sort()
print(pizza)
```

```python
pizza.sort(reverse=True)
print(pizza)
```

```python
pizza.append("portuguesa")
pizza.append("portuguesa")
pizza.append("portuguesa")
print(pizza)
```

```python
pizza.count("portuguesa")
```

```python
pizza.remove("portuguesa")
print(pizza)
```

```python
pizza.insert(3,"bode")
print(pizza)
```

```python
pizza[2]="mozzarella"
print(pizza)
```

## _Loops_ do tipo For

```python
for i in range(0, 3):
    pizza.append("a moda da casa")

print(pizza)
```

```python
for pizzas in pizza:
    print(pizzas)
```

## Loops do tipo _While_

```python
n=0
while (n<10):
    print("Oi")
    n=n+1
```

## If...else (Condicional)

```python
for i in range(1,101):
    if (i%2==0):
        print(i, "esse número é par")
    else:
        print(i, "esse número é ímpar")
```
