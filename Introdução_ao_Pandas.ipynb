### Criar Dataframe

Um *dataframe* deve ser retangular, cada coluna, também chamada de variável, deve ter o mesmo tipo de arquivo e o mesmo tamanho.

É importante que cada coluna tenha um nome fácil de lembrar e descritivo dos dados. Ex: "Nome" ao invés de "nm" ou "n". "gini" ou "indice_de_gini" ao invés de "gn".

### Parâmetros comuns de `pd.read_csv`

`pd.read_csv()` é uma função poderosa para ler dados CSV. Aqui estão alguns de seus parâmetros mais comuns e suas explicações:

*   **`filepath_or_buffer`**: O caminho para o arquivo CSV (string), URL ou um objeto com um método `read()`. Este é o único parâmetro obrigatório.

*   **`sep`** (ou `delimiter`): Caractere ou string a ser usado como separador entre os valores. O padrão é a vírgula (`,`).

*   **`header`**: Linha a ser usada como nomes das colunas. O padrão é `0` (a primeira linha). Pode ser `None` se o arquivo não tiver cabeçalho, ou uma lista de inteiros para um cabeçalho de várias linhas.

*   **`names`**: Lista de nomes de colunas a serem usados. Útil quando o arquivo não tem cabeçalho (`header=None`) ou para renomear colunas existentes.

*   **`index_col`**: Coluna a ser usada como o índice do DataFrame. Pode ser um nome de coluna ou um número inteiro. Se for `None` (o padrão), um índice numérico padrão será criado.

*   **`dtype`**: Dicionário de colunas para tipos de dados. Por exemplo, `{'coluna_x': float, 'coluna_y': str}`. Permite forçar um tipo de dado específico para uma coluna durante a leitura.

*   **`engine`**: O mecanismo de parser a ser usado. Pode ser `'c'` (mais rápido, padrão) ou `'python'` (mais flexível para alguns casos).

*   **`nrows`**: Número de linhas a serem lidas do arquivo (útil para pré-visualizar arquivos grandes).

*   **`skiprows`**: Número de linhas a pular no início do arquivo, ou uma lista de índices de linhas a pular.

*   **`na_values`**: String ou lista de strings para serem reconhecidas como valores `NaN` (valores ausentes). O padrão inclui várias strings comuns como `'NA'`, `'#N/A'`, etc.

*   **`parse_dates`**: Booleano ou lista de colunas para tentar analisar como datas. Se for `True` e não for especificado, tentará analisar o índice ou todas as colunas.

*   **`encoding`**: Codificação para usar na leitura do arquivo (e.g., `'utf-8'`, `'latin1'`).

*   **`decimal`**: Caractere a ser reconhecido como separador decimal (e.g., `','` para arquivos em português/europeu).

*   **`thousands`**: Caractere a ser reconhecido como separador de milhares (e.g., `'.'` para arquivos em português/europeu).
